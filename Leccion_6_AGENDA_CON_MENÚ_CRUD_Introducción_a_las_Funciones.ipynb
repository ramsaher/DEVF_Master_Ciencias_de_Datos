````python
lista_a = [1, 2,5,10,15]
lista_b = [3, 4, 5,100,200]

def unir_y_sumar(lista_a, lista_b):

    # Paso 1: Unimos las listas
    # Aquí 'lista_a' y 'lista_b' son los parámetros (variables de entrada)

    lista_grande = lista_a + lista_b

    # Paso 2: Usamos la función integrada sum()
    # sum() es una de las funciones integradas de Python

    total = sum(lista_grande)

    # Paso 3: Devolvemos el resultado final
    # Sin el return, la función retornaría 'None' por defecto
    return total


# Aquí pasamos los "Argumentos" (los valores reales)

resultado_final = unir_y_sumar(lista_a, lista_b)
print(resultado_final) # Esto imprimirá: 15
````

**7. Define una función para eliminar un contacto de la agenda:**

````python
# Crear función eliminar contacto de la agenda

def eliminar_contacto():
    print("\n🗑️  === Eliminar Contacto ===")
    nombre_buscar = input("  Ingresa el nombre del contacto a eliminar: ").strip().lower()
    for contacto in agenda:
        if contacto["nombre"].lower() == nombre_buscar:
            agenda.remove(contacto)
            print("✅ Contacto eliminado con éxito.")
            return
    print("⚠️  Contacto no encontrado.")

````

**8. Junta todas las funciones del menú CRUD dentro de una función:**
