```python
import numpy as np
import matplotlib.pyplot as plt
import tensorflow as tf
from tensorflow import keras
```

```python
xs = np.arange(-1,5,dtype=float)
ys = np.array([-4.0, 1.0, 6.0, 11.0, 16.0, 21.0])
print(f'xs: {xs}')
print(f'ys: {ys}')
```

```python
regresion_lineal = np.polyfit(xs, ys, 1)
beta, alfa = regresion_lineal
print(f'beta: {beta}')
print(f'alfa: {alfa}')
print(f'y = {round(beta, 1)}x + {round(alfa, 1)}')
```

```python
plt.plot(xs, ys, 'o', color = 'green', label= 'Datos')
plt.plot(xs, beta*xs + alfa, color= 'blue', label= 'RegresiГіn Lineal')
plt.axhline(y = 0, color = 'red')
plt.axvline(x = 0, color = 'red')
plt.title('Say hello to the "Hello, World" of ML')
plt.xlabel('xs')
plt.ylabel('ys')
plt.legend()
plt.grid()
plt.show()
```

```python
capa = tf.keras.layers.Dense(units=1)
modelo = tf.keras.Sequential([
    tf.keras.Input(shape=(1,)),
    capa
])

modelo.compile(optimizer='sgd', loss = 'mse')
```

```python
modelo.summary()
```

```python
print('Comenzando entrenamiento...')
historial = modelo.fit(xs, ys, epochs=500, verbose=True)
print('Modelo entrenado!')
```

# Paso 1: ConfiguraciГіn del entorno

```python
# NSTALACIГ“N Y CONFIGURACIГ“N

import numpy as np
import matplotlib.pyplot as plt
import tensorflow as tf
from tensorflow import keras
from keras import layers

print("VersiГіn de TensorFlow:", tf.__version__)
```

# Paso 2: Crear los datos

```python
# DATOS DE ENTRENAMIENTO

# Datos proporcionados en el challenge
celsius = np.array([-40.0, -10.0, 0.0, 8.0, 15.0, 22.0, 38.0], dtype=float)
fahrenheit = np.array([-40.0, 14.0, 32.0, 46.0, 59.0, 72.0, 100.0], dtype=float)

# Mostramos los datos para verificar
print("Grados Celsius:", celsius)
print("Grados Fahrenheit:", fahrenheit)
print("Cantidad de ejemplos:", len(celsius))
```

# Paso 3: Visualizar los datos

```python
# GRГҒFICO DE DISPERSIГ“N

plt.scatter(celsius, fahrenheit, color='blue', label='Datos reales')
plt.xlabel("Grados Celsius")
plt.ylabel("Grados Fahrenheit")
plt.title("RelaciГіn Celsius vs Fahrenheit")
plt.grid(True)
plt.legend()
plt.show()
```

# Paso 4: Construir la red neuronal

```python
# ARQUITECTURA DEL MODELO

modelo = keras.Sequential([
    layers.Dense(units=1, input_shape=[1])  # 1 neurona, 1 entrada (Celsius)
])

# Mostramos la estructura

modelo.summary()
```

# Paso 5: Compilar el modelo

```python
# COMPILACIГ“N

modelo.compile(
    optimizer='adam',                # Algoritmo para ajustar pesos
    loss='mean_squared_error'       # FunciГіn de pГ©rdida (error cuadrГЎtico medio)
)

print("Modelo compilado correctamente")
```

# Paso 6: Entrenar la red neuronal

```python
# ENTRENAMIENTO
print("Comenzando entrenamiento...")

historial = modelo.fit(
    celsius,            # Entradas (X)
    fahrenheit,         # Salidas reales (Y)
    epochs=500,         # Veces que verГЎ todos los datos
    verbose=1           # Muestra el progreso
)

print("Entrenamiento terminado")
```

# Paso 7: Ver la evoluciГіn de la pГ©rdida

```python
# GRГҒFICO DE PГүRDIDA POR ГүPOCA


plt.plot(historial.history['loss'], label='PГ©rdida (error)')
plt.xlabel('Гүpoca')
plt.ylabel('Error cuadrГЎtico medio')
plt.title('Comportamiento del error durante el entrenamiento')
plt.legend()
plt.grid(True)
plt.show()
```

# Paso 8: Probar el modelo con nuevos valores

```python
# PREDICCIONES

# Probamos con valores NO vistos durante el entrenamiento
celsius_prueba = np.array([-20.0, 10.0, 30.0, 100.0], dtype=float)
fahrenheit_predicho = modelo.predict(celsius_prueba)

print("Predicciones del modelo:")
for c, f in zip(celsius_prueba, fahrenheit_predicho):
    print(f"{c}В°C = {f[0]:.2f}В°F (predicho)")

# Comparamos con la fГіrmula real
print("\nComparaciГіn con fГіrmula real:")
for c in celsius_prueba:
    real = c * 1.8 + 32
    print(f"{c}В°C = {real}В°F (real)")
```

# Paso 9: Ver los pesos aprendidos

```python
# INSPECCIГ“N DEL MODELO

peso = modelo.layers[0].get_weights()[0][0][0]
sesgo = modelo.layers[0].get_weights()[1][0]

print("Peso (pendiente) aprendido:", peso)
print("Sesgo (intercepto) aprendido:", sesgo)
print("\nFГіrmula de la neurona:")
print(f"F = C * {peso:.4f} + {sesgo:.4f}")
print("\nFГіrmula real:")
print("F = C * 1.8 + 32")
```

# Paso 10: Guardar el modelo

```python
# GUARDAR MODELO

modelo.save("mymodel.h5")
print("Modelo guardado como 'mymodel.h5'")

# Verificar que existe
import os
if os.path.exists("mymodel.h5"):
    print("вң… Archivo guardado correctamente")
    print("TamaГұo:", os.path.getsize("mymodel.h5"), "bytes")
```

# Paso 11 Cargar y probar el modelo guardado

```python
# VERIFICAR CARGA DEL MODELO

# Cerramos TensorFlow y lo volvemos a cargar
# (Esto simula que abrimos el modelo despuГ©s de guardarlo)

modelo_cargado = keras.models.load_model("mymodel.h5")
print("Modelo cargado desde disco")

# Probamos nuevamente

celsius_nuevo = np.array([25.0, 37.0], dtype=float)
fahrenheit_nuevo = modelo_cargado.predict(celsius_nuevo)

for c, f in zip(celsius_nuevo, fahrenheit_nuevo):
    print(f"{c}В°C = {f[0]:.2f}В°F (modelo cargado)")
```
