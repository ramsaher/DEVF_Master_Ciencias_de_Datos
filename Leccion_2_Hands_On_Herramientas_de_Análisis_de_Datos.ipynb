### **SERIES**

Una **Serie** de Pandas, es un objeto similar a un vector, que contiene una secuencia de valores y un vector asociado de etiquetas llamado indice.

Basicamente, una serie es una columna de un DataFrame.

De manera similar a los arrays de Numpy, podemos crear una serie pasandole una lista a su constructor.

````python
# Crear array vector con 4 elementos secuenciales del 1 al 4

vector_array = np.arange(1, 5)

# np.arange(1, 5) crea números desde 1 hasta 4 (el 5 no se incluye)
# Resultado: [1, 2, 3, 4]
````

````python
vector_array
````

````python
# Crear una serie de números con los elementos del 1 al 4

serie_simple = pd.Series([1, 2, 3, 4])

# pd.Series() convierte una lista en una Serie de Pandas
# Cada elemento tendrá un índice automático: 0, 1, 2, 3

print("Serie simple:")
print(serie_simple)
````

Adicional a las funciones de creacion de datos de Numpy, Pandas tambien tiene funcionalidad de creacion de datos, por ejemplo podemos hacer un indice con un rango de datos usando `pd.date_range`.

````python
# Crear DatetimeIndex con fecha de inicio del año 2024 con 4 periodos de frecuencia por mes


fechas = pd.date_range(start = '2024', periods = 4, freq = 'YS') # freq = 's'(segundo), 'min'(minuto), 'h'(hora), 'd'(día), 'W'(semana),
fechas                                                           # freq = 'MS'(inicio de mes), 'ME'(fin de mes),
                                                                 #'freq = 'QS'(inicio de trimestre), 'QE'(fin de trimestre),
                                                                 # freq = 'YS'(inicio de año), 'YE'(fin de año)


print("Frecuencia diaria (periods=5)")
fechas_diarias = pd.date_range(start='2024-01-01', periods=5, freq='D')
print(fechas_diarias)

print("Frecuencia semanal (periods=3)")
fechas_semanales = pd.date_range(start='2024-01-01', periods=3, freq='W')
print(fechas_semanales)

print("Frecuencia de inicio de trimestre (periods=4)")
fechas_trimestrales = pd.date_range(start='2024-01-01', periods=4, freq='QS-JAN') # QS-JAN significa trimestre que comienza en enero
print(fechas_trimestrales)

print("Frecuencia horaria (periods=3)")
fechas_horarias = pd.date_range(start='2024-01-01 10:00', periods=3, freq='H')
print(fechas_horarias)

````

````python
print("\n--- Ejemplos de 'start' ---")
print("El parámetro 'start' puede ser una cadena de fecha/hora en varios formatos o un objeto Timestamp.")

print("'start' como 'YYYY-MM-DD'")
start_str_1 = pd.date_range(start='2023-03-15', periods=3, freq='D')
print(start_str_1)

print("'start' como 'MM/DD/YYYY'")
start_str_2 = pd.date_range(start='03/15/2023', periods=3, freq='D')
print(start_str_2)

print("'start' con hora y minuto 'YYYY-MM-DD HH:MM'")
start_str_3 = pd.date_range(start='2023-03-15 09:30', periods=3, freq='H')
print(start_str_3)

print("'start' como un objeto Timestamp")
start_timestamp = pd.Timestamp('2023-03-15 14:00:00')
fechas_timestamp = pd.date_range(start=start_timestamp, periods=3, freq='D')
print(fechas_timestamp)
````

````python
# Crear DatetimeIndex con fecha de inicio del año 2024 con 4 periodos de frecuencia por mes

fechas = pd.date_range(start = '2024', periods = 4, freq = 'YS') # freq = 's'(segundo), 'min'(minuto), 'h'(hora), 'd'(día), 'W'(semana),
fechas                                                           # freq = 'MS'(inicio de mes), 'ME'(fin de mes),
                                                                 #'freq = 'QS'(inicio de trimestre), 'QE'(fin de trimestre),
                                                                 # freq = 'YS'(inicio de año), 'YE'(fin de año)

````

A diferencia de los Numpy arrays, los indices son arbitrarios y nos permite acceder relacionar la informacion con algun dato de interes. Ademas nos permite seleccionar la informacion por medio del uso de estos indices.

````python
# Crear serie con los elementos del array vector y con los indices de la serie de fechas

serie_fechas = pd.Series(vector_array, index=fechas)

# Asignamos los valores [1,2,3,4] a las fechas que creamos

print("Serie con fechas como índice:")
print(serie_fechas)
````

````python
# Renombrar columna por la letra A

serie_fechas.name = 'A'

# .name asigna un nombre a la Serie (como el título de una columna)

print("Serie renombrada:")
print(serie_fechas)
````

#### **INDEXACIÓN**

La **indexación** en Pandas, similar a NumPy y listas de Python, permite acceder a elementos por posición. Sin embargo, Pandas destaca por sus índices explícitos que van mas allá de la numérica, posibilitando la selección basada en etiquetas además de la posición.

````python
# Indexar serie por indice de fecha

fecha_especifica = '2024-02-29'
valor = serie_fechas[fecha_especifica]
print(f"Valor en {fecha_especifica}: {valor}")
# Puedes acceder a un valor usando su "etiqueta" (la fecha)
````

````python
# Indexar serie por indices de fechas salteadas

fechas_salteadas = ['2024-01-31', '2024-03-31']
valores_salteados = serie_fechas[fechas_salteadas]
print("Valores en fechas específicas:")
print(valores_salteados)

# Pasamos una lista de etiquetas y obtenemos una nueva Serie
````

````python
# Indexar serie por condición booleana

condicion = serie_fechas > 2

# Esto crea una Serie de True/False: ¿cada valor es mayor a 2?

valores_mayores_2 = serie_fechas[condicion]
print("Valores mayores a 2:")
print(valores_mayores_2)

# Filtramos: solo mostramos donde la condición es True
````

````python
# Indexar serie por rango de indice

primeros_dos = serie_fechas[0:2]
print("Primeros dos elementos:")
print(primeros_dos)

# [0:2] significa: desde índice 0 hasta antes del índice 2
````

### **DATAFRAMES**

Un **DataFrame** de Pandas representa una tabla rectangular de datos y contiene una colección ordenada de columnas, cada una de las cuales puede ser un tipo de valor diferente (numérico, string, booleano, etc).

Los DataFrames tienen un índice de fila y columna; se puede pensar en
como un diccionario de **Series**, las cuales comparten el mismo índice.

Una manera de crear un DataFrames es a partir de las Series ya creadas en Python.

````python
# Crear dataframe haciendo uso de un diccionario

df = pd.DataFrame({
    'A': [1, 2, 3, 4],
    'B': [5, 6, 7, 8],
    'C': [9, 10, 11, 12],
    'D': [13, 14, 15, 16]
})
# Un diccionario donde cada clave es el nombre de una columna
# Cada valor es una lista con los datos de esa columna

print("DataFrame creado:")
print(df)
````

````python
# Crear un DataFrame desde un diccionario

datos = {
    "nombre": ["Ana", "Luis", "Frank", "Gemi", "Pete"],
    "edad": [14, 22, 33, 44, 55],
    "ciudad": ["Madrid", "Barcelona", "Valencia", "Sevilla", "Bilbao"],
    "activo": [True, False, False, True, True]
}

df = pd.DataFrame(datos)
df
````

````python
# Cambiar los indices de la filas del dataframe por una serie de fecha

df.index = pd.date_range(start='2024-01-01', periods=5, freq='ME') # periods=5

# Asígnamos fechas como identificadores de filas

print("DataFrame con índices de fecha:")
print(df)
````

El atributo `dtypes` nos permite conocer el tipo de datos de cada columna de un DataFrame

````python
# Observar los tipos de datos de cada columna del dataframe

print("Tipos de datos por columna:")
print(df.dtypes)

# int64 significa números enteros
# float64 significa números decimales
# object significa texto
````

Para extrae el indice de la tabla se utiliza `.index`  

````python
# Extraer los indices del dataFrame

print("Índices del DataFrame:")
print(df.index)

# Muestra las etiquetas de las filas
````

Para extraer las columnas de la tabla se utiliza `.columns`  

````python
# Extraer las columnas del dataFrame

print("Columnas del DataFrame:")
print(df.columns)

# Muestra los nombres de las columnas
````

#### **INDEXACIÓN**

Los **DataFrames** comparten el mismo índice que las Series. Esto permite una indexación potente y flexible, similar a las Series, pero en dos dimensiones (filas y columnas).

````python
# Indexar dataframe por variable e indice

columna_nombre = df['nombre']

# Los corchetes con una cadena de texto seleccionan una columna

print("Columna 'nombre':")
print(columna_nombre)
````

````python
# Indexar dataframe por condición booleana en la variable


condicion_df = df['edad'] > 30

# ¿Qué filas tienen la edad mayor a 30?

filtro_df = df[condicion_df]
print("Filas donde edad > 30:")
print(filtro_df)
````

````python
# Indexar dataframe por indices y variables conjuntas

valor_especifico = df.loc['2024-02-29', 'edad']

# .loc usa ETIQUETAS (nombres)

print(f"Valor en fila 2024-02-29, columna edad: {valor_especifico}")
````

````python
# Indexar dataframe por indices y variables salteadas

subconjunto = df.loc[['2024-01-31', '2024-03-31'], ['nombre', 'ciudad']]

# Seleccionamos filas específicas y columnas específicas

print("Subconjunto de datos:")
print(subconjunto)
````

##### **INDEXACIÓN POR MÉTODOS EXPECIALES**


En Pandas, `.loc` e `.iloc` son **métodos fundamentales para acceder a datos** dentro de un DataFrame.

*   `.loc` se basa en etiquetas, lo que te permite seleccionar filas y columnas por sus nombres o una condición booleana.


*   `.iloc` utiliza índices numéricos, proporcionando acceso a los datos según su posición entera en el DataFrame.

**Ambos métodos son esenciales para filtrar, seleccionar y manipular datos de manera eficiente** en Pandas, aunque la elección entre ellos depende de prefirir usar etiquetas o posiciones numéricas para identificar los elementos de interes.

###### **MÉTODO POR ETIQUETA**

````python
# Indexar dataframe por indices y variables salteadas


subconjunto_loc = df.loc[['2024-01-31', '2024-03-31'], ['nombre', 'ciudad']]
print("Usando .loc (por etiquetas):")
print(subconjunto_loc)
````

````python
# Mostrar el elemento en una fecha y columna específicas del dataframe

# El DataFrame 'df' actual tiene índices de fecha y columnas como 'nombre', 'edad', 'ciudad', 'activo'.
# Para acceder a un elemento, debemos usar un índice de fecha y un nombre de columna válidos.

elemento_seleccionado = df.loc['2024-01-31', 'nombre']
print(f"Elemento en la fila '2024-01-31', columna 'nombre': {elemento_seleccionado}")

elemento_seleccionado_2 = df.loc['2024-03-31', 'edad']
print(f"Elemento en la fila '2024-03-31', columna 'edad': {elemento_seleccionado_2}")
````

````python
# Indexar dataframe por condición booleana en la variable

condicion_loc = df['edad'] > 30 # Usamos 'edad' para la condición
filtro_loc = df.loc[condicion_loc, ['nombre', 'ciudad']] # Usamos 'nombre' y 'ciudad' para la selección
print("Usando .loc con condición:")
print(filtro_loc)
````

###### **MÉTODO POR POSICIÓN**

````python
# Indexar dataframe por indices y variables salteadas

subconjunto_iloc = df.iloc[[0, 2], [0, 2]]
# iloc usa POSICIONES numéricas (fila 0, fila 2, columna 0, columna 2)
print("Usando .iloc (por posición):")
print(subconjunto_iloc)
````

````python

datos_grandes = {
    'Col1': range(1, 16),
    'Col2': [x * 2 for x in range(1, 16)],
    'Col3': [x * 3 for x in range(1, 16)],
    'Col4': [x * 4 for x in range(1, 16)],
    'Col5': [f'Valor_{x}' for x in range(1, 16)]
}
df_large = pd.DataFrame(datos_grandes)

print("DataFrame grande creado (df_large):")
print(df_large)
````

### **TIPOS DE GRÁFICOS**

Seaborn ofrece una variedad de tipos de gráficos para visualizar datos. Muchos de ellos muestran el **análisis de comportamientos de una o dos variables**.

#### **GRÁFICO DE DISPERSIÓN**

Un gráfico de dispersión se utiliza para visualizar la relación o correlación entre dos variables numéricas continuas. Cada punto en el gráfico representa una observación individual, con su posición determinada por los valores de las dos variables (una en el eje X y la otra en el eje Y).

````python
# Gráfico de dispersión que muestra la relación entre la cuenta total y la propina

plt.figure(figsize=(10, 6))  # Tamaño del gráfico (ancho, alto)

sns.scatterplot(data=tips, x='total_bill', y='tip')
plt.title('Relación entre Cuenta Total y Propina')
plt.xlabel('Cuenta Total ($)')
plt.ylabel('Propina ($)')
plt.show()

# scatterplot = puntos en un plano
# x = eje horizontal (total_bill)
# y = eje vertical (tip)
# Cada punto es una comida/transacción
````

#### **HISTOGRAMA**

Un histograma es una representación gráfica de la distribución de una única variable numérica. Divide el rango de valores de la variable en una serie de intervalos o "bins" (generalmente de igual tamaño) y luego cuenta cuántas observaciones caen en cada intervalo. La altura de cada barra en el histograma es proporcional a la frecuencia (o densidad) de observaciones en ese intervalo.

````python
# Histograma que muestra la distribución de la cuenta total

plt.figure(figsize=(10, 6))
sns.histplot(data=tips, x='total_bill', bins=20, kde=True)

# bins=20: número de barras (intervalos)
# kde=True: añade una curva suave que muestra la densidad

plt.title('Distribución de Cuentas Totales')
plt.xlabel('Cuenta Total ($)')
plt.ylabel('Frecuencia')
plt.show()

# El histograma Nos dice: "¿qué valores son más comunes?"
````

#### **GRÁFICO DE BARRAS**

Los gráficos de barras se utilizan para comparar una métrica numérica (como una media, suma o conteo) entre diferentes categorías de una variable cualitativa o discreta. Cada barra representa una categoría, y la altura (o longitud) de la barra corresponde al valor de la métrica numérica para esa categoría. Seaborn, por defecto, suele mostrar la media de la variable numérica para cada categoría, junto con un intervalo de confianza.

````python
# Gráfico de barras que muestra la propina promedio por día de la semana

plt.figure(figsize=(10, 6))
sns.barplot(data=tips, x='day', y='tip', palette='viridis')

# barplot = barras, muestra el promedio por defecto
# day: categorías (jueves, viernes, sábado, domingo)
# tip: el valor a promediar

plt.title('Propina Promedio por Día de la Semana')
plt.xlabel('Día')
plt.ylabel('Propina Promedio ($)')
plt.show()
````

#### **GRÁFICO DE CAJA**

Un gráfico de caja, también conocido como diagrama de caja y bigotes, es una forma estandarizada de mostrar la distribución de datos numéricos basada en cinco números resumen: el mínimo, el primer cuartil (Q1, percentil 25), la mediana (Q2, percentil 50), el tercer cuartil (Q3, percentil 75) y el máximo. La "caja" representa el rango intercuartílico (IQR = Q3 - Q1), que contiene el 50% central de los datos. Los "bigotes" se extienden desde la caja para mostrar el resto de la distribución, típicamente hasta 1.5 veces el IQR. Los puntos más allá de los bigotes se consideran valores atípicos.

````python
# Gráfico de caja nos muestra la distribución de la cuenta total por hora del día

plt.figure(figsize=(10, 6))
sns.boxplot(data=tips, x='time', y='total_bill', palette='Set2')

# time: Lunch (almuerzo) o Dinner (cena)
# El boxplot muestra: mediana, cuartiles, valores atípicos

plt.title('Distribución de Cuentas por Momento del Día')
plt.xlabel('Momento')
plt.ylabel('Cuenta Total ($)')
plt.show()
````

#### **GRÁFICO DE VIOLÍN**

Un gráfico de violín es una combinación de un gráfico de caja y un gráfico de densidad (Kernel Density Estimate o KDE) rotado y colocado a cada lado para mostrar la forma de la distribución de los datos. La parte central del "violín" puede mostrar un mini gráfico de caja o simplemente la mediana y los cuartiles. El ancho del violín en un punto particular representa la densidad de los datos en ese valor.

````python
# Gráfico de violín que nos muestra la distribución de la propina por sexo

plt.figure(figsize=(10, 6))
sns.violinplot(data=tips, x='sex', y='tip', palette='muted')

# Violin = caja + densidad (muestra la "forma" de los datos)

plt.title('Distribución de Propinas por Género')
plt.xlabel('Género')
plt.ylabel('Propina ($)')
plt.show()
````

#### **MAPA DE CALOR**

Un mapa de calor es una representación gráfica de datos donde los valores individuales contenidos en una matriz se representan como colores.1 Generalmente, se utiliza una escala de colores para indicar la magnitud de los valores; por ejemplo, colores más oscuros o más intensos pueden representar valores más altos, y colores más claros, valores más bajos (o viceversa). Las filas y columnas de la matriz suelen representar diferentes variables o categorías.

````python
# Convierte las columnas 'sexo' y 'fumador' a representaciones numéricas

tips_encoded = tips.copy()  # Copiamos para no modificar original
tips_encoded['sex_encoded'] = tips_encoded['sex'].map({'Male': 0, 'Female': 1})
tips_encoded['smoker_encoded'] = tips_encoded['smoker'].map({'Yes': 1, 'No': 0})
tips_encoded['day_encoded'] = tips_encoded['day'].map({'Thur': 0, 'Fri': 1, 'Sat': 2, 'Sun': 3})
tips_encoded['time_encoded'] = tips_encoded['time'].map({'Lunch': 0, 'Dinner': 1})

# .map() transforma texto a números (necesario para correlación)

print("Primeras filas con datos codificados:")
print(tips_encoded[['sex', 'sex_encoded', 'smoker', 'smoker_encoded']].head())
````

````python
# Mapa de calor que muestra la correlación entre las variables numéricas

# Seleccionamos solo columnas numéricas

columnas_numericas = ['total_bill', 'tip', 'size', 'sex_encoded', 'smoker_encoded', 'day_encoded', 'time_encoded']
correlacion = tips_encoded[columnas_numericas].corr()

plt.figure(figsize=(12, 8))
sns.heatmap(correlacion, annot=True, cmap='coolwarm', center=0)

# heatmap = mapa de calor
# annot=True: muestra los números dentro de las celdas
# cmap='coolwarm': colores (rojo=correlación positiva, azul=negativa)

plt.title('Mapa de Calor de Correlaciones')
plt.show()

# ¿Qué es correlación? Mide si dos variables se mueven juntas
# Cercano a 1: van en la misma dirección
# Cercano a -1: van en direcciones opuestas
# Cercano a 0: no hay relación
````

### **PERSONALIZACIÓN DE GRÁFICOS**

Seaborn se basa en Matplotlib, lo que te permite personalizar aún más tus gráficos. Puedes usar las funciones de Matplotlib para agregar títulos, etiquetas, leyendas, cambiar colores, tamaños de fuente y mucho más.

**3. Creación de métrica: Porcentaje de Propina**

* Crea una nueva columna llamada tip_percentage.
* Calcular el porcentaje entre la propina y el total de la cuenta.
* Visualiza los primeros 10 registros.

````python
# Creación de columna calculada

tips['tip_percentage'] = (tips['tip'] / tips['total_bill']) * 100

# Fórmula: (propina / cuenta total) × 100 = porcentaje de propina

print("="*50)
print("PRIMEROS 10 REGISTROS CON PORCENTAJE DE PROPINA")
print("="*50)
print(tips[['total_bill', 'tip', 'tip_percentage']].head(10))
print("\n")

# Estadísticas del nuevo indicador

print("="*50)
print("ESTADÍSTICAS DEL PORCENTAJE DE PROPINA")
print("="*50)
print(f"Porcentaje promedio de propina: {tips['tip_percentage'].mean():.2f}%")
print(f"Porcentaje mínimo de propina: {tips['tip_percentage'].min():.2f}%")
print(f"Porcentaje máximo de propina: {tips['tip_percentage'].max():.2f}%")
print(f"Desviación estándar: {tips['tip_percentage'].std():.2f}%")
````

**5. Análisis Visual: ¿Cuánto gastan los fumadores?**

* Utiliza un gráfico de dispersión para ver la relación entre la cuenta total y la propina.

* Utiliza el parámetro `hue` para diferenciar a los clientes que son fumadores de los que no.

* Personaliza las dimensiones y los colores del gráfico para una mejor visualización con `style`.

````python
# Realizar gráfico de dispersión

plt.figure(figsize=(12, 7))


scatter_plot = sns.scatterplot(
    data=tips,
    x='total_bill',
    y='tip',
    hue='smoker',      # Color por fumador/no fumador
    style='smoker',    # Estilo de marcador diferente
    size='size',       # Tamaño según comensales
    sizes=(50, 250),   # Rango de tamaños
    palette={'Yes': '#FF6B6B', 'No': '#4ECDC4'},  # Colores personalizados
    alpha=0.7          # Transparencia (permite ver puntos superpuestos)
)

# Personalización del gráfico con Matplotlib

plt.title('Relación entre Gasto y Propina: Comparativa Fumadores vs No Fumadores',
          fontsize=14, fontweight='bold')
plt.xlabel('Cuenta Total ($)', fontsize=12)
plt.ylabel('Propina ($)', fontsize=12)

# Añadir línea de tendencia (propina del 15%)

x_line = np.array([0, 60])
y_line_15 = x_line * 0.15  # 15% de propina
y_line_20 = x_line * 0.20  # 20% de propina

plt.plot(x_line, y_line_15, '--', color='gray', alpha=0.5, label='Propina 15%')
plt.plot(x_line, y_line_20, '--', color='darkgray', alpha=0.5, label='Propina 20%')

plt.legend(title='¿Fumador?', title_fontsize='10', loc='upper left')
plt.grid(True, alpha=0.3)

# Añadir anotación con estadísticas

propina_promedio_fumadores = tips[tips['smoker'] == 'Yes']['tip_percentage'].mean()
propina_promedio_no_fumadores = tips[tips['smoker'] == 'No']['tip_percentage'].mean()

plt.text(45, 8, f'Fumadores: {propina_promedio_fumadores:.1f}% promedio',
         fontsize=9, style='italic')
plt.text(45, 7.5, f'No Fumadores: {propina_promedio_no_fumadores:.1f}% promedio',
         fontsize=9, style='italic')

plt.tight_layout()
plt.show()

````

**6. Análisis de Frecuencia: Días de mayor afluencia**

* Utiliza un gráfico de conteo para visualizar qué días de la semana el restaurante recibe más mesas.  

* Utiliza el parámetro `hue` para diferenciar a los días conforme al conteo de mensas que son fumadores de los que no.

* Personaliza las dimensiones y los colores del gráfico para una mejor visualización. Se recomienda utilizar `palette='viridis'` para una mejor experiencia visual respecto al color.

````python
# Realizar gráfico de barras

plt.figure(figsize=(12, 6))

# Creamos gráfico de conteo

count_plot = sns.countplot(
    data=tips,
    x='day',
    hue='smoker',  # Diferenciamos por fumadores
    palette='viridis',  # Paleta de colores recomendada
    order=['Thur', 'Fri', 'Sat', 'Sun'],  # Orden lógico de días
    edgecolor='black'  # Borde de las barras
)

# Personalización

plt.title('Afluencia de Clientes por Día de la Semana', fontsize=14, fontweight='bold')
plt.xlabel('Día de la Semana', fontsize=12)
plt.ylabel('Número de Mesas / Transacciones', fontsize=12)

# Rótulos más descriptivos en el eje X

dias_completos = {
    'Thur': 'Jueves',
    'Fri': 'Viernes',
    'Sat': 'Sábado',
    'Sun': 'Domingo'
}
count_plot.set_xticklabels([dias_completos[item.get_text()] for item in count_plot.get_xticklabels()])

# Añadir valores sobre las barras

for container in count_plot.containers:
    count_plot.bar_label(container, fontsize=10, fontweight='bold')

plt.legend(title='¿Fumador?', title_fontsize='10', labels=['No Fumador', 'Fumador'])
plt.grid(True, alpha=0.3, axis='y')

plt.tight_layout()
plt.show()

# Análisis adicional: porcentaje de fumadores por día

print("\n" + "="*50)
print("ANÁLISIS ADICIONAL: PORCENTAJE DE FUMADORES POR DÍA")
print("="*50)

for dia in ['Thur', 'Fri', 'Sat', 'Sun']:
    df_dia = tips[tips['day'] == dia]
    total_dia = len(df_dia)
    fumadores_dia = len(df_dia[df_dia['smoker'] == 'Yes'])
    porcentaje = (fumadores_dia / total_dia) * 100
    print(f"{dias_completos[dia]}: {porcentaje:.1f}% de fumadores ({fumadores_dia}/{total_dia})")
````
