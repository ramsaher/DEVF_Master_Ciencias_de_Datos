### **Gestion de una pequeña biblioteca personal.**

```python
# Nuestro lista es un "estante" donde guardaremos los libros (registros)
# Esta lista es nuestra base de datos mientras el programa corre
# Cada diccionario es un libro

biblioteca = [
    {"id": 1, "titulo": "El Principito", "autor": "Antoine de Saint-Exupéry", "año": 1943, "precio": 129.50},
    {"id": 2, "titulo": "Cien Años de Soledad", "autor": "Gabriel García Márquez", "año": 1967, "precio": 210.00},
    {"id": 3, "titulo": "Don Quijote de la Mancha", "autor": "Miguel de Cervantes", "año": 1605, "precio": 185.00},
    {"id": 4, "titulo": "1984", "autor": "George Orwell", "año": 1949, "precio": 185.00},
    {"id": 5, "titulo": "El resplandor", "autor": "Stephen King", "año": 1977, "precio": 210.00},
    {"id": 6, "titulo": "Crónicas de una muerte anunciada", "autor": "Gabriel García Márquez", "año": 1981, "precio": 145.00},
    {"id": 7, "titulo": "El Código Da Vinci", "autor": "Dan Brown", "año": 2003, "precio": 289.00},
    {"id": 8, "titulo": "Orgullo y Prejuicio", "autor": "Jane Austen", "año": 1813, "precio": 195.50},
    {"id": 9, "titulo": "La Tregua", "autor": "Mario Benedetti", "año": 1960, "precio": 175.00},
    {"id": 10, "titulo": "Frankenstein", "autor": "Mary Shelley", "año": 1818, "precio": 130.00}
]
```

```python
   # Función para visualizar los libros

def ver_todos():

    # Nos muestra los libros en formato de tabla y el total al final

    print("\n  Inventario de la Biblioteca  ")

    print("\n" + "="*65)

    # Los encabezados con formato de espacios: <5 (izquierda), <35 (centro/izq), etc.

    print(f" {'ID':<4} | {'TÍTULO':<35} | {'AÑO':<6} | {'PRECIO':<10}")
    print("="*65)

    if not biblioteca:
        print("La biblioteca está vacía.")
    else:
        for libro in biblioteca:

    # f-string con formato de moneda y alineación

            print(f" {libro['id']:<4} | {libro['titulo']:<35} | {libro['año']:<6} | $ {libro['precio']:>8.2f}")

    print("="*65)

    # Función integrada len() para contar los elementos de la lista

    print(f" 📚 Total de libros: {len(biblioteca)}")




```

```python
# Función para crear registro nuevo de libro

def crear_registro():

#  Crea un nuevo libro solicitando todos los campos requeridos

    print("\n--- Registrar Nuevo Libro ---")
    try:

# Generamos un ID automáticamente basado en el último registro

        nuevo_id = biblioteca[-1]["id"] + 1 if biblioteca else 1

        titulo = input("Título: ")
        autor = input("Autor: ")
        año = int(input("Año de publicación: "))
        precio = float(input("Precio: "))

        libro = {
            "id": nuevo_id,
            "titulo": titulo,
            "autor": autor,
            "año": año,
            "precio": precio
        }

        biblioteca.append(libro)
        print(f"✅ Libro '{titulo}' registrado con ID {nuevo_id}.")
    except ValueError:
        print("⚠️ Error: El año y el precio deben ser números.")
```

```python
# Función para consultar  libro o registro


def consultar_registro():

# Busca un libro por su nombre exacto

    ver_todos()
    if not biblioteca: return
    busqueda = input("\nEscribe el título del libro que buscas: ")
    encontrado = False

    for libro in biblioteca:
        if libro["titulo"].lower() == busqueda.lower():
            print(f"🔍 Resultado: El autor de '{libro['titulo']}' es {libro['autor']}.")
            encontrado = True
            break

    if not encontrado:
        print("❌ No encontramos ese libro.")



# Función  para cambiar los datos de un libro existente

def actualizar_registro():

    ver_todos()
    if not biblioteca: return

    try:
        indice = int(input("\nEscribe el número del libro a editar: ")) - 1
        nuevo_titulo = input("Nuevo título (deja vacío para no cambiar): ")
        nuevo_autor = input("Nuevo autor (deja vacío para no cambiar): ")

  # Selección: Si el usuario escribió algo, lo cambiamos

        if nuevo_titulo:
            biblioteca[indice]["titulo"] = nuevo_titulo
        if nuevo_autor:
            biblioteca[indice]["autor"] = nuevo_autor

        print("✨ Registro actualizado.")
    except:
        print("⚠️ Error: Elige un número válido de la lista.")


# Función para eliminar registro

def eliminar_registro():

# Saca un libro de la lista

    ver_todos()
    if not biblioteca: return

    try:
        indice = int(input("\nEscribe el número del libro que quieres borrar: ")) - 1

# Usamos pop para eliminar por posición

        eliminado = biblioteca.pop(indice)
        print(f"🗑️ Se ha eliminado '{eliminado['titulo']}'.")
    except:
        print("⚠️ No se pudo eliminar. Revisa el número.")
```
